```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted", font_scale=1.1)
```

```python
# Load dataset directly from seaborn
df = sns.load_dataset("titanic")

print("âœ… Dataset loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
print("\nðŸ“Š Dataset Info:")
df.info()

print("\nðŸ§ª Summary Statistics:")
display(df.describe(include='all').transpose())
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap="viridis")
plt.title("Missing Values in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="survived", data=df, palette="Set2")
plt.title("Survival Count (0=No, 1=Yes)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="survived", hue="sex", data=df, palette="Set1")
plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["age"].dropna(), bins=30, kde=True, color="skyblue")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x="pclass", y="fare", data=df, palette="Pastel1")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns

plt.figure(figsize=(6,4))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="pclass", y="survived", hue="sex", data=df, palette="Set2")
plt.title("Survival Rate by Class and Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
embarked_survived = df.groupby(["embarked", "survived"]).size().unstack()
embarked_survived.plot(kind="bar", stacked=True, figsize=(6,4), color=["red","green"])
plt.title("Survival Count by Port of Embarkation")
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.show()
```
